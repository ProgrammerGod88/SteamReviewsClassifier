```python
from langdetect import detect
```

```python
def removeNonEnglish(Review_Text):
    try:
        language = detect(Review_Text)
        if language == "en":
            return True  
        else:
            return False  
    except Exception as e:
        return False
    

English_Reviews = dota_reviews[dota_reviews["review"].apply(removeNonEnglish)]
    
```
